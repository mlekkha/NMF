# Tutorial 1: Getting Started with the NMF Repository

This tutorial gets you from a blank computer to having **Jupyter Lab** open with the NMF example project ready to run. It is written for readers who are new to the command line, `git`, and Python package management — every command you need to type is shown, along with the output you should expect to see.

By the end of this tutorial you will have:

1. Installed **GitHub Desktop** and used it to clone the NMF repository to your computer.
2. Installed **Miniforge3**, a lightweight distribution of the `conda` package manager.
3. Created the `NMF` conda **environment**, which contains the exact versions of Python, NumPy, scikit-learn, and other packages the code needs.
4. Launched **Jupyter Lab** and opened the notebook for [Tutorial 2](../example/Repele2605/Tutorial2_RunningNMF.ipynb), where you will actually run the NMF analysis.

> **Note on how to use this document.** Steps 1–2 (installing software, cloning the repository, running terminal commands) happen *before* Jupyter Lab exists on your computer, so you can't run them from inside a notebook. Read this tutorial as a PDF or in your web browser on GitHub, and type the commands into a Terminal window yourself. Only once you reach Step 6 will you actually be inside Jupyter.

## What you'll need

- A Mac or Linux computer (Windows users: install [Windows Subsystem for Linux](https://learn.microsoft.com/windows/wsl/install) first, then follow the Linux instructions below inside it).
- An internet connection and about 5 GB of free disk space.
- A free [GitHub](https://github.com/join) account.
- Permission to install software on the computer (administrator/sudo access).

Estimated time: 20–30 minutes, most of which is waiting for downloads.

## Step 1: Install GitHub Desktop

[GitHub Desktop](https://desktop.github.com) is a free application with a graphical interface (no typing required) for downloading and managing copies of GitHub repositories.

1. Go to **https://desktop.github.com** in your web browser.
2. Click **Download for macOS** (or **Download for Windows**).
3. Open the downloaded file and follow the installer's instructions (drag the app into your Applications folder on macOS, or run the installer on Windows).
4. Open GitHub Desktop. The first time it runs, click **Sign in to GitHub.com** and log in with your GitHub account.

> **Linux users:** GitHub does not publish an official Linux build of GitHub Desktop. You have to clone the repository from the command line instead (skip to the box below).
>
> ```bash
> git clone https://github.com/mlekkha/NMF.git
> ```
>
> This creates an `NMF` folder in your current directory containing the same files GitHub Desktop would have downloaded. If you use this method, skip ahead to **Step 3**.

## Step 2: Clone the NMF repository with GitHub Desktop

"Cloning" means downloading a full copy of a GitHub repository, including its entire history, to your computer.

1. In GitHub Desktop, choose **File > Clone Repository...** from the menu bar.
2. Click the **URL** tab.
3. Paste in the repository URL: `https://github.com/mlekkha/NMF`
4. Under **Local Path**, choose (or leave the default) folder where the repository will be saved — for example `~/Documents/GitHub/NMF` on macOS. Make a note of this path; you will need it later.
5. Click **Clone**.

**Expected result:** GitHub Desktop shows a progress bar while it downloads the repository, then displays the NMF repository with its file history. If you open the folder shown under **Local Path** in Finder (macOS) or File Explorer (Windows), you should see:

```text
NMF/
├── GeneData.py
├── LICENSE
├── NMF.yaml
├── README.md
├── example/
│   └── Repele2605/
│       ├── performNMFOnRepele2605Data.ipynb
│       ├── Repele2605RNASeqTimeSeries.csv
│       └── ... (other data files)
└── tutorials/
    ├── Tutorial1_GettingStarted.ipynb
    └── ...
```

## Step 3: Open a terminal

The remaining steps use the **command line** (also called the "terminal" or "shell") — a text-based way of giving your computer instructions. Every command below is meant to be typed (or copy-pasted) into this window, followed by pressing **Enter**.

- **macOS:** Press `Cmd+Space` to open Spotlight, type `Terminal`, and press Enter.
- **Linux:** Open your distribution's terminal application (often found in the applications menu, or via `Ctrl+Alt+T`).

You should see a window with a prompt that ends in `$` or `%`, waiting for you to type something.

## Step 4: Install Miniforge3

The NMF code depends on specific versions of Python and several scientific-computing packages (NumPy, scikit-learn, Matplotlib, JupyterLab). **Conda** is a tool that installs and manages self-contained bundles of software called *environments*, so that these packages don't conflict with anything else on your computer. **Miniforge3** is the small, free distribution of conda we'll use.

In your terminal, type the following two commands, pressing Enter after each one:

```bash
curl -L -O "https://github.com/conda-forge/miniforge/releases/latest/download/Miniforge3-$(uname)-$(uname -m).sh"
bash Miniforge3-$(uname)-$(uname -m).sh
```

The first command downloads the installer for your specific computer type (macOS vs. Linux, Intel vs. Apple Silicon/ARM). The second runs it.

**Expected output** while the installer runs — it will pause and wait for you to respond at a few points:

```text
Welcome to Miniforge3 24.x.x

In order to continue the installation process, please review the license
agreement.
Please, press ENTER to continue
>>> [press Enter]
... (license text) ...
Do you accept the license terms? [yes|no]
[no] >>> yes

Miniforge3 will now be installed into this location:
/Users/yourname/miniforge3

  - Press ENTER to confirm the location
  - Press CTRL-C to abort the installation
  - Or specify a different location below

>>> [press Enter]
PREFIX=/Users/yourname/miniforge3
Unpacking payload ...
Installing base environment...
...
do you wish to update your shell profile to automatically initialize conda?
This will activate conda on startup and change the command prompt when activated.
[yes|no]
[no] >>> yes
```

Type `yes` at both prompts (license agreement and shell profile update), pressing Enter after each.

After it finishes, **close the terminal window and open a new one** (or run `source ~/.bash_profile` / `source ~/.zshrc`) so that the changes take effect. Then verify the install worked:

```bash
conda --version
```

```text
conda 24.3.0
```

(Your version number may differ slightly — that's fine.)

## Step 5: Create the NMF conda environment

The repository includes a file called `NMF.yaml` that lists exactly which packages the code needs. First, move into the folder where GitHub Desktop cloned the repository (use the path you noted in Step 2):

```bash
cd ~/Documents/GitHub/NMF
```

Then create the environment from the recipe file:

```bash
conda env create -f NMF.yaml
```

This reads `NMF.yaml`, downloads the listed packages, and installs them into a new environment named `NMF`. It can take a few minutes depending on your internet connection.

**Expected output:**

```text
Channels:
 - bioconda
 - conda-forge
 - nodefaults
Platform: osx-arm64
Collecting package metadata (repodata.json): ...working... done
Solving environment: ...working... done

Downloading and Extracting Packages: ...working... done
Preparing transaction: ...working... done
Verifying transaction: ...working... done
Executing transaction: ...working... done
#
# To activate this environment, use
#
#     $ conda activate NMF
#
# To deactivate an active environment, use
#
#     $ conda deactivate
```

(You may also see a one-time notice that a newer version of conda is available — you can safely ignore it, or run the suggested `conda update` command later.)

Now activate the environment:

```bash
conda activate NMF
```

Your terminal prompt should now start with `(NMF)`, showing the environment is active. Verify the key packages installed correctly:

```bash
python -c "import sklearn, numpy, matplotlib; print('Packages OK')"
```

```text
Packages OK
```

## Troubleshooting

| Problem | Fix |
|---|---|
| `conda: command not found` after installing Miniforge3 | Close and reopen your terminal window, or run `source ~/miniforge3/etc/profile.d/conda.sh`. |
| Terminal prompt doesn't show `(NMF)` when you expect it to | Run `conda activate NMF` again — the environment needs to be activated in every new terminal window. |
| `jupyter lab` says the port is already in use | Jupyter will automatically try the next port (8889, 8890, ...) and print the new URL — just use that one. You can also specify one yourself with `jupyter lab --port=8889`. |
| GitHub Desktop isn't available for Linux | See the note in Step 1 — use https://github.com/shiftkey/desktop, or clone with `git clone https://github.com/mlekkha/NMF.git` from the terminal. |
| `conda env create` fails partway through | Make sure you're in the `NMF` folder (run `pwd` to check) and that `NMF.yaml` exists there (run `ls`), then try again. |
